```python
import ast
import pandas as pd

def safe_literal_eval(x):
    if pd.isna(x):
        return []
    if isinstance(x, (list, dict)):
        return x
    try:
        return ast.literal_eval(x)
    except:
        return []

for col in ["genres", "keywords", "cast", "crew"]:
    movies_df[col] = movies_df[col].apply(safe_literal_eval)
```

```python
# Step 2: Convert stringified lists to actual Python lists
def safe_eval(x):
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return []
```

```python
# Convert list columns
list_columns = ['genres', 'keywords', 'cast', 'crew']
for col in list_columns:
    movies_df[col] = movies_df[col].apply(safe_eval)
```

```python
initial_count = len(movies_df)
movies_df = movies_df.drop_duplicates(subset=['movie_id'], keep='first')
```

```python
final_count = len(movies_df)
final_count
```

```python
movies_df = movies_df.dropna(subset=['title', 'genres', 'cast', 'director'])
print(f"  Final count: {len(movies_df)} movies")
```

```python
def clean_token(token):
    """Clean individual tokens"""
    token = str(token).strip()
    # Remove special characters
    token = ''.join(c for c in token if c.isalnum() or c == ' ')
    # Remove extra spaces
    token = ' '.join(token.split())
    return token.lower()
```

```python
for col in list_columns:
    movies_df[col] = movies_df[col].apply(
        lambda x: [clean_token(item) for item in x if item]
    )
```

```python
for col in list_columns:
    movies_df[col] = movies_df[col].apply(lambda x: [item for item in x if item])
```

```python
def create_tokens(row):
    """
    Combine all categorical fields into single token list
    """
    tokens = []
    
    tokens.extend(row['genres'])
    
    tokens.extend(row['keywords'])
    
    cast_list = row['cast']
    tokens.extend(cast_list[:3] if len(cast_list) >= 3 else cast_list)
    
    if pd.notna(row['director']) and row['director']:
        tokens.append(clean_token(row['director']))
    
    if pd.notna(row['release_year']):
        tokens.append(f"year{int(row['release_year'])}")
    
    seen = set()
    unique_tokens = []
    for token in tokens:
        if token not in seen:
            seen.add(token)
            unique_tokens.append(token)
    
    return unique_tokens
```

```python
movies_df['tokens'] = movies_df.apply(create_tokens, axis=1)
```

```python
# token statistics
token_counts = movies_df['tokens'].apply(len)
print(f"\nToken count statistics:")
print(f"  Min tokens per movie: {token_counts.min()}")
print(f"  Max tokens per movie: {token_counts.max()}")
print(f"  Avg tokens per movie: {token_counts.mean():.1f}")
print(f"  Median tokens per movie: {token_counts.median():.1f}")
```

```python
w2v_model_path = "../artifacts/word2vec_model.model"
w2v_emb_path = "../artifacts/word2vec_movie_embeddings.npy"
```

```python
if os.path.exists(w2v_model_path):
    print(f"\nLoading existing model from '{w2v_model_path}'")
    w2v_model = Word2Vec.load(w2v_model_path)
    print(f"  Vocabulary size: {len(w2v_model.wv)}")
else:
    print(f"\nTraining new Word2Vec model...")
    print(f"  Total sentences: {len(sentences)}")
    
    w2v_model = Word2Vec(
        sentences=sentences,
        vector_size=100,        # Embedding dimension
        window=5,               # Context window size
        min_count=2,            # Minimum word frequency
        workers=4,              # Parallel threads
        epochs=10,              # Training iterations
        sg=1,                   # Skip-gram algorithm
        seed=42                 # For reproducibility
    )
    
    # Save the model
    w2v_model.save(w2v_model_path)
    print(f"  Vocabulary size: {len(w2v_model.wv)}")
```

```python
# Show similar words to "action"
print("\nWords similar to 'action':")
try:
    similar_action = w2v_model.wv.most_similar('action', topn=5)
    for word, score in similar_action:
        print(f"  {word}: {score:.4f}")
except:
    print("  'action' not in vocabulary")
```

```python
# Show similar words to "thriller"
print("\nWords similar to 'thriller':")
try:
    similar_thriller = w2v_model.wv.most_similar('thriller', topn=5)
    for word, score in similar_thriller:
        print(f"  {word}: {score:.4f}")
except:
    print("  'thriller' not in vocabulary")
```

```python
# Generate Movie Embeddings
def get_movie_embedding(tokens, model, vector_size=100):
    vectors = []
    for token in tokens:
        if token in model.wv:
            vectors.append(model.wv[token])
    
    if len(vectors) == 0:
        return np.zeros(vector_size)
    
    return np.mean(vectors, axis=0)
```

```python
if os.path.exists(w2v_emb_path):
    print(f"\n Loading existing embeddings from '{w2v_emb_path}'")
    word2vec_embeddings = np.load(w2v_emb_path)
    print(f"  Shape: {word2vec_embeddings.shape}")
else:
    print(f"\n Generating embeddings for {len(sentences)} movies.")
    
    word2vec_embeddings = np.array([
        get_movie_embedding(tokens, w2v_model) 
        for tokens in sentences
    ])
    
    # Save embeddings
    np.save(w2v_emb_path, word2vec_embeddings)
    print(f" Embeddings generated and saved!")
    print(f"  Shape: {word2vec_embeddings.shape}")
```

```python
def recommend_movies_w2v(query, top_n=10):

    # Tokenize and clean query
    query_tokens = query.lower().split()
    query_tokens = [token.strip() for token in query_tokens if token.strip()]
    
    # Get query embedding
    query_emb = get_movie_embedding(query_tokens, w2v_model)
    
    # Load embeddings
    embeddings = np.load(w2v_emb_path)
    
    # Calculate similarity
    scores = cosine_similarity([query_emb], embeddings)[0]
    
    # Get top N indices
    idx = np.argsort(scores)[::-1][:top_n]
    
    # Return recommendations
    recommendations = movies_df.iloc[idx][[
        'title', 'genres', 'cast', 'director', 
        'vote_average', 'release_year', 'score'
    ]].copy()
    
    recommendations['similarity_score'] = scores[idx]
    recommendations = recommendations.reset_index(drop=True)
    
    return recommendations
```

## Recommendation using Transformer (SBERT)

```python
# Convert string-form list → Python list
def parse_list(x):
    if isinstance(x, list):
        return x
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return [x]
    return [x]
```

```python
# Create movie description with all relevant features
def generate_description(row):
    title = row["title"]
    year = row["release_year"]  
    
    overview = row["overview"]
    genres = ", ".join(parse_list(row["genres"]))
    keywords = ", ".join(parse_list(row["keywords"]))
    cast = ", ".join(parse_list(row["cast"]))
    
    director = row["director"] if row["director"] else "Unknown Director"
    
    language = row["original_language"]
    score = row["score"]

    return (
        f"{title} ({year}) is a {genres} film in {language} language directed by {director}. "
        f"It stars {cast}. "
        f"The storyline is: {overview}. "
        f"Important themes and keywords include: {keywords}. "
        f"Movie score: {score:.2f}."
    )
```

```python
movies["description"] = movies.apply(generate_description, axis=1)
```

```python
import os
import numpy as np
from sentence_transformers import SentenceTransformer

emb_path = "../artifacts/recommendation_embeddings.npy"

model = SentenceTransformer("all-MiniLM-L6-v2")

# Check if embedding file exists
if os.path.exists(emb_path):
    print(f"Embedding file found at {emb_path}")
    recommendation_embeddings = np.load(emb_path)
    print(f"Embeddings loaded!")
else:
    print(f"No embedding file found. Generating embeddings from descriptions.")
    
    # Generate embeddings
    recommendation_embeddings = model.encode(
        movies["description"].tolist(),
        show_progress_bar=True,
        batch_size=32  
    )
    
    np.save(emb_path, recommendation_embeddings)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_movies(query, top_n=10):
    
    # Encode query using the same model
    q_emb = model.encode([query])

    # Load saved embeddings (use consistent path)
    recommendation_embeddings = np.load("../artifacts/recommendation_embeddings.npy")

    # Calculate cosine similarity between query and all movies
    scores = cosine_similarity(q_emb, recommendation_embeddings)[0]

    # Get indices of top N highest similarity scores
    idx = np.argsort(scores)[::-1][:top_n]

    # Create recommendations dataframe with additional columns
    recommendations = movies.iloc[idx][["title", "genres", "cast", "vote_average", "release_year", "director", "score"]].copy()
    
    # Add similarity score to show relevance
    recommendations['similarity_score'] = scores[idx]
    
    # Reset index for cleaner output
    recommendations = recommendations.reset_index(drop=True)
    
    return recommendations
```
